First stage of the Maslow's pyramid is Physiological needs. Lets check together that how many years last to buy an 50m2 apartment in Turkey!

```python
data = "/kaggle/input/global-cost-of-living/cost-of-living_v2.csv"
Ph = pd.read_csv(data)
```

```python
Ph.info()
```

```python
Ph.head()
```

```python
Ph = Ph[["city","country","x1","x9","x10","x11","x12","x13","x19","x20","x21","x31","x36","x37","x38","x48","x49","x52","x53","x54"]]
```

```python
Ph.columns=["City","Country","Meal","Milk","Bread","Rice","Eggs","Cheese","Tomato","Potato","Onion","Taxi","Basic","Mobile","Internet","Rent Center","Rent Outside","Apt Center","Apt Outside","Salary"]
```

```python
Ph.head(3)
```

Monthly spend for per person who lives outside of the city

```python
Ph["Monthly Spend"] = Ph["Meal"] + Ph["Milk"] + Ph["Bread"] + Ph["Rice"] + Ph["Eggs"] + Ph["Cheese"] + Ph["Tomato"] + Ph["Potato"] + Ph["Onion"] + Ph["Taxi"] + Ph["Basic"] + Ph["Mobile"] + Ph["Internet"] + Ph["Rent Outside"]
```

```python
Ph.head(3)
```

```python
Ph1 = Ph[["City", "Country", "Apt Outside", "Salary", "Monthly Spend"]]
```

```python
Ph1
```

```python
Ph2 = Ph1.dropna()
```

```python
Ph2 = Ph2.copy()
```

```python
Ph2.info()
```

```python
Ph2["50 m2"] = Ph2["Apt Outside"]*50
```

Years that a person be able to buy an apartment outside the city? (assuming zero inflation)

```python
Ph2["Years"] = Ph2["50 m2"]/(Ph2["Salary"] - Ph2["Monthly Spend"])/12
```

```python
Ph2.head(3)
```

```python
Tr = Ph2[Ph2["Country"] == "Turkey"]
```

```python
Tr.sort_values(by = "Years")
```

```python
Tr["Years"].mean()
```
